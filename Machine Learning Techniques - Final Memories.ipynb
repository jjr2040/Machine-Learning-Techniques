**¿Que es una red CNN?**

La CNN es un tipo de Red Neuronal Artificial con aprendizaje supervisado que procesa sus capas imitando al cortex visual del ojo humano para identificar diferentes características en las entradas, las cuales hacen que pueda identificar cualquier objeto. ### **3. Arquitectura de la red**

![Arquitectura CNN](https://raw.githubusercontent.com/alejitro/proyecto_MLT_Grupo05/master/red-neuronal-convolucional-arquitectura.png)

Una red neuronal convolucional (CNN) es una red multicapa que consta de una capa de entrada, capas convolucionales, capas de reducción y capas de conexión total, como por ejemplo, una red perceptrón multicapa.

**Capa de Entrada:** Son los pixeles de la imagen. Es el alto, ancho y profundidad. Se podrá utilizar un sólo color o tres para Red,Green,Blue.


**Capa de Convolución:** En la convolución se realizan operaciones de productos y sumas entre la capa de entrada y los n filtros (o kernel) que genera un mapa de características. Los características extraídas corresponden a cada posible ubicación del filtro en la imagen original.

Después de aplicar la convolución se le aplica a los mapas de características una función de activación. Generalmente, la función de activación que se aplica a los elementos de la matriz es ReLU


**Capa de Reducción:** También conocida como POOL ó SUBSAMPLING, en esta capa se disminuye la cantidad de parámetros al quedarse con las características más comunes. Es decir, hará una reducción en las dimensiones alto y ancho, pero se mantendrá la profundidad.


**Capa de conexión total o clasificadora:** En el final de las capas convolucionales y de reducción, se suele utilizar capas completamente conectadas en la que cada pixel se considera como una neurona separada al igual que en un perceptrón multicapa. La última capa de esta red es una capa clasificadora que tendrá tantas neuronas como el número de clases a predecir.

#### Sobremuestreo de imagenes

Hay 253 imagenes de resonancia magnetica, de las cuales 155 contienen tumores y 90 no contienen tumores
**About the data:
The dataset contains 2 folders: yes and no which contains 253 Brain MRI Images. The folder yes contains 155 Brain MRI Images that are tumorous andno contains 98 Brain MRI Images that are non-tumorous.**

````python
# se define eldirectorio con los datos 
image_dir="input/brain-mri-images-for-brain-tumor-detection/"
#se crean los directorios para las imagenes spbremuestreadas
os.makedirs('../output/kaggle/working/augmented-images')
os.makedirs('../output/kaggle/working/augmented-images/yes')
os.makedirs('../output/kaggle/working/augmented-images/no')

# se define la función que va a generar las nuevas imagenes a partir de las ya existentes
def augment_data(file_dir, n_generated_samples, save_to_dir):
    data_gen = ImageDataGenerator(rotation_range=10, 
                                  width_shift_range=0.1, 
                                  height_shift_range=0.1, 
                                  shear_range=0.1, 
                                  brightness_range=(0.3, 1.0),
                                  horizontal_flip=True, 
                                  vertical_flip=True, 
                                  fill_mode='nearest'
                                 )

    for filename in listdir(file_dir):
        image = cv2.imread(file_dir + '/' + filename)
        # reshape the image
        image = image.reshape((1,)+image.shape)
        save_prefix = 'aug_' + filename[:-4]
        i=0
        for batch in data_gen.flow(x=image, batch_size=1, save_to_dir=save_to_dir,save_prefix=save_prefix, save_format='jpg'):
                i += 1
                if i > n_generated_samples:
                    break
                    
augmented_data_path ='../output/kaggle/working/augmented-images/'
# generamos datos para la clase positiva
augment_data(file_dir=image_dir+'yes',n_generated_samples=6, save_to_dir=augmented_data_path+'yes')
# se generan datos para la clase negativa, se generan un mayor numero de muestras
augment_data(file_dir=image_dir+'no', n_generated_samples=9, save_to_dir=augmented_data_path+'no')

````

#### Procesamiento de los datos

Se recotan los bordes de la imagen

````python
def crop_brain_contour(image, plot=False):
    
    # Convert the image to grayscale, and blur it slightly
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    gray = cv2.GaussianBlur(gray, (5, 5), 0)
    
    thresh = cv2.threshold(gray, 45, 255, cv2.THRESH_BINARY)[1]
    thresh = cv2.erode(thresh, None, iterations=2)
    thresh = cv2.dilate(thresh, None, iterations=2)

    # Find contours in thresholded image, then grab the largest one
    cnts = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    cnts = imutils.grab_contours(cnts)
    c = max(cnts, key=cv2.contourArea)
    # extreme points
    extLeft = tuple(c[c[:, :, 0].argmin()][0])
    extRight = tuple(c[c[:, :, 0].argmax()][0])
    extTop = tuple(c[c[:, :, 1].argmin()][0])
    extBot = tuple(c[c[:, :, 1].argmax()][0])
    
    # crop new image out of the original image using the four extreme points (left, right, top, bottom)
    new_image = image[extTop[1]:extBot[1], extLeft[0]:extRight[0]]            

    if plot:
        plt.figure()
        plt.subplot(1, 2, 1)
        plt.imshow(image)
        plt.tick_params(axis='both', which='both', top=False, bottom=False, left=False, right=False,labelbottom=False, labeltop=False, labelleft=False, labelright=False)
        plt.title('Original Image')
        plt.subplot(1, 2, 2)
        plt.imshow(new_image)
        plt.tick_params(axis='both', which='both',top=False, bottom=False, left=False, right=False,labelbottom=False, labeltop=False, labelleft=False, labelright=False)
        plt.title('Cropped Image')
        plt.show()
    
    return new_image
````

#### Carga de los datos para el modelo

````python
def load_data(dir_list, image_size):

    # load all images in a directory
    X = []
    y = []
    image_width, image_height = image_size
    
    for directory in dir_list:
        for filename in listdir(directory):
            image = cv2.imread(directory+'/'+filename)
            image = crop_brain_contour(image, plot=False)
            image = cv2.resize(image, dsize=(image_width, image_height), interpolation=cv2.INTER_CUBIC)
            # normalize values
            image = image / 255.
            # convert image to numpy array and append it to X
            X.append(image)
            # append a value of 1 to the target array if the image
            # is in the folder named 'yes', otherwise append 0.
            if directory[-3:] == 'yes':
                y.append([1])
            else:
                y.append([0])
                
    X = np.array(X)
    y = np.array(y)
    
    # Shuffle the data
    X, y = shuffle(X, y)
    
    print(f'Number of examples is: {len(X)}')
    print(f'X shape is: {X.shape}')
    print(f'y shape is: {y.shape}')
    
    return X, y

augmented_yes =augmented_data_path+'yes'
augmented_no = augmented_data_path+'no'

IMG_WIDTH, IMG_HEIGHT = (240, 240)

X, y = load_data([augmented_yes, augmented_no], (IMG_WIDTH, IMG_HEIGHT))
````

#### Visualización de muestras de ejemplo

````python
#funcion que sirve para dibujar un numero n de muestras
def plot_sample_images(X, y, n=40):
    for label in [0,1]:
        # grab the first n images with the corresponding y values equal to label
        images = X[np.argwhere(y == label)]
        n_images = images[:n]
        
        columns_n = 10
        rows_n = int(n/ columns_n)

        plt.figure(figsize=(10, 8))
        
        i = 1 # current plot        
        for image in n_images:
            plt.subplot(rows_n, columns_n, i)
            plt.imshow(image[0])
            
            # remove ticks
            plt.tick_params(axis='both', which='both', 
                            top=False, bottom=False, left=False, right=False,
                           labelbottom=False, labeltop=False, labelleft=False, labelright=False)
            
            i += 1
        
        label_to_str = lambda label: "Yes" if label == 1 else "No"
        plt.suptitle(f"Brain Tumor: {label_to_str(label)}")
        plt.show()
````

````python
plot_sample_images(X, y)
````

#### Particionamiento de los datos en test & train

````python
def split_data(X, y, test_size=0.2):
       
    X_train, X_test_val, y_train, y_test_val = train_test_split(X, y, test_size=test_size)
    X_test, X_val, y_test, y_val = train_test_split(X_test_val, y_test_val, test_size=0.5)
    
    return X_train, y_train, X_val, y_val, X_test, y_test


X_train, y_train, X_val, y_val, X_test, y_test = split_data(X, y, test_size=0.3)

print ("cantidad de observaciones en training = " + str(X_train.shape[0]))
print ("cantidad de observaciones para validación= " + str(X_val.shape[0]))
print ("cantidad de observaciones para test = " + str(X_test.shape[0]))
````

#### Construcción del modelo 

````python
def build_model(input_shape):
    X_input = Input(input_shape) 
    X = ZeroPadding2D((2, 2))(X_input) 
    
    X = Conv2D(32, (7, 7), strides = (1, 1))(X)
    X = BatchNormalization(axis = 3, name = 'bn0')(X)
    X = Activation('relu')(X) 
    
    X = MaxPooling2D((4, 4))(X) 
    X = MaxPooling2D((4, 4))(X) 
    X = Flatten()(X) 
    X = Dense(1, activation='sigmoid')(X) 
    model = Model(inputs = X_input, outputs = X)
    
    return model
````

````python
IMG_SHAPE = (IMG_WIDTH, IMG_HEIGHT, 3)
model=build_model(IMG_SHAPE)
model.summary()
````

#### Entrenamiento del modelo 

````python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(x=X_train, y=y_train, batch_size=30, epochs=22, validation_data=(X_val, y_val))
````

````python
history = model.history.history
````

#### Evaluación del modelo

````python
def plot_metrics(history):
    
    train_loss = history['loss']
    val_loss = history['val_loss']
    train_acc = history['accuracy']
    val_acc = history['val_accuracy']
    
    # Loss
    plt.figure()
    plt.plot(train_loss, label='Training Loss')
    plt.plot(val_loss, label='Validation Loss')
    plt.title('Loss')
    plt.legend()
    plt.show()
    
    # Accuracy
    plt.figure()
    plt.plot(train_acc, label='Training Accuracy')
    plt.plot(val_acc, label='Validation Accuracy')
    plt.title('Accuracy')
    plt.legend()
    plt.show()
````

````python
plot_metrics(history)
````

Las ESN parten del computo del reservorio, que tiene las siguientes características:
<ul>
<li>Los pesos entre la capa de entrada y la capa oculta se asignan aleatoriamente y no son entrenables. </li>
<li>Los pesos de las neuronas de salida son entrenables y se pueden aprender para que la red pueda reproducir patrones temporales específicos. </li>
<li>La capa oculta (o el reservorio) está escasamente conectada (menor al 10% de conectividad)</li>
<li>La arquitectura del reservorio crea una incrustación no lineal recurrente (h en la Ilustración 1) de la entrada que luego se puede conectar a la salida deseada y estos pesos finales serán entrenables.</li>
<li>Se puede conectar a un modelo predictivo diferente.</li>


### **1. Razones para utilizar la red**

Las arquitecturas tradicionales de redes neuronales sufren del problema de explosión del gradiente, el cual consiste en una acumulación grande de gradientes de error que resulta en actualizaciones muy grandes para los pesos de los modelos de redes neuronales en entrenamiento. Esto hace que el modelo sea inestable e incapaz de aprender los datos de entrenamiento. Las Echo State Networks (ESN) no sufren de este problema.<br><br>
Otra de las razones es que las ESN son muy rápidas, ya que no hay fase de propagación hacia atrás en el reservorio mientras que las redes neuronales tradicionales son computacionalmente costosas.<br><br>
Las ESN están bien adaptadas para manejar series de tiempo caóticas y para no ser interrumpidas por bifucarciones, todo lo contrario a las redes neuronales tradicionales.

Se usan `w^in`, `W`, `w^out` como matrices de los respectivos pesos de conexión sináptica. Utilizando `f(x)=tanh⁡ x` como salida no linealidad para todas las unidades de capa oculta. 

Las principales diferencias de ESN con respecto a los enfoques tradicionales de redes recurrentes son la configuración de los pesos de conexión y el procedimiento de entrenamiento. Para construir un ESN, las unidades en la capa de entrada y la capa oculta están conectados al azar. Las conexiones entre la capa oculta y las unidades de salida son las únicas conexiones que se entrenan, generalmente con un enfoque de aprendizaje supervisado y fuera de línea que utiliza la regresión lineal. Sin embargo, para que el enfoque funcione con éxito, las conexiones en el depósito no pueden ser completamente aleatorias; Los *reservoirs* de ESN están típicamente diseñados para tener la propiedad de estado de eco (ESP). 

Para entrenar una ESN se siguen los siguientes pasos:

1. Proporcionar un RNN aleatorio: Debe crear un *reservoir* dinámico aleatorio RNN, utilizando cualquier modelo de neurona y el tamaño de él depende de la tarea. Luego, se debe adjuntar las unidades de entrada al *reservoir* creando conexiones aleatorias de todo a todos. Por último en este paso, se debe crear unidades de salida, si la tarea requiere retroalimentación de salida se instala conexiones de salida a depósito generadas aleatoriamente (todo a todo). Pero si la tarea no requiere retroalimentación, no se crea ninguna conexión.

2. Cosechar los estados del *reservoir*: Conduce el depósito dinámico con los datos de entrenamientore por tiempos 

3. Calcular los pesos de salida: Se calculan los pesos de salida como en regresión lineal, luego estos pesos se usan para crear conexiones de *reservoir*. Al terminar esto, el ESN esta listo para usarse. 

Referencia: [Information Processing in Echo State Networks at the Edge of Chaos](http://lizier.me/joseph/publications/2010-Boedecker-InfoProcessingEchoStateNets.pdf)

### **3. Arquitectura de la red**

Los GANs son una de las ideas más interesantes del mundo del Machine Learning en la actualidad. Dos modelos son entrenados simultaneamente, de manera que un modelo (generador) aprende a crear datos que parecen reales mientras otro (discriminador) aprende a distinguir entre datos reales y falsos.

![Arqutectura](https://user-images.githubusercontent.com/63893101/89554409-6eca1f00-d7d4-11ea-97d3-54a85b0a8f79.png)

Las GANs realizan una comparacion indirecta entre los modelos, que toma la forma de una tarea posterior sobre estas dos distribuciones. El entrenamiento de la red generativa se realiza con respecto a esta tarea de tal manera que obliga a la distribución generada a acercarse más y más a la distribución verdadera.

![funcionamiento](https://user-images.githubusercontent.com/63893101/89555826-2449a200-d7d6-11ea-8de2-b55ae1cc6d77.gif)

La tarea posterior de las GAN es una tarea de discriminación entre muestras verdaderas y generadas. Entonces, en una arquitectura GAN, tenemos un discriminador, que toma muestras de datos verdaderos y generados y que trata de clasificarlos lo mejor posible, y un generador capacitado para engañar al discriminador tanto como sea posible.El generador es una red neuronal que modela una función de transformación. Toma como entrada una variable aleatoria simple y debe devolver, una vez entrenada, una variable aleatoria que sigue la distribución objetivo. Como es muy complicado y desconocido, decidimos modelar el discriminador con otra red neuronal. Esta red neuronal modela una función discriminativa. Toma como entrada un punto (en nuestro ejemplo de perro un vector N dimensional) y devuelve como salida la probabilidad de que este punto sea "verdadero".

![funcionamiento](https://user-images.githubusercontent.com/63893101/89556153-96ba8200-d7d6-11ea-80cf-3beae1a62515.png)

En resumen, el proceso de modelamiento se requiere esencialmente definir dos cosas: la arquitectura y la funcion de costo. Ya hemos descrito previamente la arquitectura. Esta consiste en dos redes neuronales: 
1. Una red generadora G(.) que toma una entrada aleatoria z con densidad p_z retorna una salida x_g = G(z) que debe seguir la distribucion objetivo. 

2. Una red discriminadora D(.) que toma una entrada x que puede ser real o falsa y retorna la probabilidad D(x) de ser un dato real.

Este método cuantifica que tan bien el discriminador es capaz de distinguir entre imagenes reales y falsas. Compara las predicciones del discriminador con los arreglos de imagenes reales (1s) y los arreglos de imagenes falsas (0s).

```
def discriminator_loss(real_output, fake_output):
    real_loss = cross_entropy(tf.ones_like(real_output), real_output)
    fake_loss = cross_entropy(tf.zeros_like(fake_output), fake_output)
    total_loss = real_loss + fake_loss
    return total_loss
```

```
@tf.function
def train_step(images):
    noise = tf.random.normal([BATCH_SIZE, noise_dim])

    with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
      generated_images = generator(noise, training=True)

      real_output = discriminator(images, training=True)
      fake_output = discriminator(generated_images, training=True)

      gen_loss = generator_loss(fake_output)
      disc_loss = discriminator_loss(real_output, fake_output)

    gradients_of_generator = gen_tape.gradient(gen_loss, generator.trainable_variables)
    gradients_of_discriminator = disc_tape.gradient(disc_loss, discriminator.trainable_variables)

    generator_optimizer.apply_gradients(zip(gradients_of_generator, generator.trainable_variables))
    discriminator_optimizer.apply_gradients(zip(gradients_of_discriminator, discriminator.trainable_variables))
```

### **5. Propiedades e hiperparametros**

Los GANs consisten en dos modelos entrenados de manera simultanea para desarrollar tareas de generador y discriminador (como previamente se explicó). Por lo tanto, los hiperparametros dependen de los modelos utilizados para dichas tareas. Aquí algunos de los hiperparametros de las redes neuronales (que son los modelos más utilizados):

**Hiperparametros asociados a la estructura de la red**

**1. Numero de capas ocultas:** Las capas ocultas son las capas entre la entrada y la salida de la red. Muchas capas ocultas con una tecnica de regularización incrementan la exactitud.  Un pequeño numero de capas puede causar un bajoajuste.

**2. Dropout:** Es una técnica de de regularización que se utiliza para evitar el sobreajuste y aumentar el poder de generalización de los modelos generados. Generalmente se usan valores de dropout de 20-50%. Es especialmente util cuando se tienen redes grandes dotando al modelo de la capacidad de aprender diferentes representaciones independientes. 

**3. Funcion de activacion:**  Las funciones de activacion son utilizadas para introducir no linealidad a los modelos, lo cual permite que los modelos de deep learning aprendan limites de decision no lineales

**Hiperparametros asociados al entrenamiento del modelo**

**1. Tasa de aprendizaje:** La tasa de aprendizaje define que tan rápido la red actualiza sus parametros. Una baja tasa de aprendizaje relentiza el aprendizaje pero converge mejor. Tasas de aprendizaje más grandes pueden acelerarlo pero se pierde convergencia. Usualmente una tasa de aprendizaje decadente es preferida.

**2. Momentum:** El momentum ayuda a saber la dirección del proximo paso con el conocimiento adquirido en el paso previo, lo cual ayuda a prevenir oscilaciones. Una tipica eleccion del momentum es entre 0.5 y 0.9

**3. Numero de epocas:** Es el numero de veces que la data de entrenamiento completa recorre la red durante el entrenamiento. Se recomienda incrementar el numero de epocas hasta que la exactitud del modelo sea optima, incluso cuando hay sobreajuste.

**3. Batch sizes:** Es el numero de submuestras dadas a la red, después de lo cual ocurre la actualización de los parametros. 

**NOTA:** Todos los parametros antes mencionados son sujetos de optimización. Para más detalles de este proceso ir al siguiente [link](https://machinelearningmastery.com/grid-search-hyperparameters-deep-learning-models-python-keras/).

````python
#Import the libraries we will need.

from __future__ import print_function, division

from keras.datasets import mnist
from keras.layers import Input, Dense, Reshape, Flatten, Dropout
from keras.layers import BatchNormalization, Activation, ZeroPadding2D
from keras.layers.advanced_activations import LeakyReLU
from keras.layers.convolutional import UpSampling2D, Conv2D
from keras.models import Sequential, Model
from keras.optimizers import Adam

import matplotlib.pyplot as plt

import sys

import numpy as np
````

2. Se realiza la estandarización de las caracteristicas de las imagenes 

````python
img_rows = 28
img_cols = 28
channels = 1
latent_dim = 100
img_shape = (img_rows, img_cols, channels)
````

3. Se inicializan los modelos, tanto el generador como el discriminador

````python
def build_generator():

    model = Sequential()

    model.add(Dense(128 * 7 * 7, activation="relu", input_dim=latent_dim))
    model.add(Reshape((7, 7, 128)))
    model.add(UpSampling2D())
    model.add(Conv2D(128, kernel_size=3, padding="same"))
    model.add(BatchNormalization(momentum=0.8))
    model.add(Activation("relu"))
    model.add(UpSampling2D())
    model.add(Conv2D(64, kernel_size=3, padding="same"))
    model.add(BatchNormalization(momentum=0.8))
    model.add(Activation("relu"))
    model.add(Conv2D(channels, kernel_size=3, padding="same"))
    model.add(Activation("tanh"))

    model.summary()

    noise = Input(shape=(latent_dim,))
    img = model(noise)

    return Model(noise, img)
````

````python
def build_discriminator():

    model = Sequential()

    model.add(Conv2D(32, kernel_size=3, strides=2, input_shape=img_shape, padding="same"))
    model.add(LeakyReLU(alpha=0.2))
    model.add(Dropout(0.25))
    model.add(Conv2D(64, kernel_size=3, strides=2, padding="same"))
    model.add(ZeroPadding2D(padding=((0,1),(0,1))))
    model.add(BatchNormalization(momentum=0.8))
    model.add(LeakyReLU(alpha=0.2))
    model.add(Dropout(0.25))
    model.add(Conv2D(128, kernel_size=3, strides=2, padding="same"))
    model.add(BatchNormalization(momentum=0.8))
    model.add(LeakyReLU(alpha=0.2))
    model.add(Dropout(0.25))
    model.add(Conv2D(256, kernel_size=3, strides=1, padding="same"))
    model.add(BatchNormalization(momentum=0.8))
    model.add(LeakyReLU(alpha=0.2))
    model.add(Dropout(0.25))
    model.add(Flatten())
    model.add(Dense(1, activation='sigmoid'))

    model.summary()

    img = Input(shape=img_shape)
    validity = model(img)

    return Model(img, validity)
````

4. Se establece el optimizador que se va a utilizar, se construyen y se compilan los modelos

````python
optimizer = Adam(0.0002, 0.5)

# Build and compile the discriminator
discriminator = build_discriminator()
discriminator.compile(loss='binary_crossentropy',
    optimizer=optimizer,
    metrics=['accuracy'])

# Build the generator
generator = build_generator()

# The generator takes noise as input and generates imgs
z = Input(shape=(latent_dim,))
img = generator(z)

# For the combined model we will only train the generator
discriminator.trainable = False

# The discriminator takes generated images as input and determines validity
valid = discriminator(img)

# The combined model  (stacked generator and discriminator)
# Trains the generator to fool the discriminator
combined = Model(z, valid)
combined.compile(loss='binary_crossentropy', optimizer=optimizer)
````

Se agreaga ruido a las imagenes para mejorar los resultados del entrenamiento


````python
def save_imgs(epoch):
    r, c = 5, 5
    noise = np.random.normal(0, 1, (r * c, latent_dim))
    gen_imgs = generator.predict(noise)

    # Rescale images 0 - 1
    gen_imgs = 0.5 * gen_imgs + 0.5

    fig, axs = plt.subplots(r, c)
    cnt = 0
    for i in range(r):
        for j in range(c):
            axs[i,j].imshow(gen_imgs[cnt, :,:,0], cmap='gray')
            axs[i,j].axis('off')
            cnt += 1
    fig.savefig("images/mnist_%d.png" % epoch)
    plt.close()
````

Se construye la función de entrenamiento

````python
def train(epochs, batch_size=128, save_interval=50):

    # Load the dataset
    (X_train, _), (_, _) = mnist.load_data()

    # Rescale -1 to 1
    X_train = X_train / 127.5 - 1.
    X_train = np.expand_dims(X_train, axis=3)

    # Adversarial ground truths
    valid = np.ones((batch_size, 1))
    fake = np.zeros((batch_size, 1))

    for epoch in range(epochs):

        # ---------------------
        #  Train Discriminator
        # ---------------------

        # Select a random half of images
        idx = np.random.randint(0, X_train.shape[0], batch_size)
        imgs = X_train[idx]

        # Sample noise and generate a batch of new images
        noise = np.random.normal(0, 1, (batch_size, latent_dim))
        gen_imgs = generator.predict(noise)

        # Train the discriminator (real classified as ones and generated as zeros)
        d_loss_real = discriminator.train_on_batch(imgs, valid)
        d_loss_fake = discriminator.train_on_batch(gen_imgs, fake)
        d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)

        # ---------------------
        #  Train Generator
        # ---------------------

        # Train the generator (wants discriminator to mistake images as real)
        g_loss = combined.train_on_batch(noise, valid)

        # Plot the progress
        print ("%d [D loss: %f, acc.: %.2f%%] [G loss: %f]" % (epoch, d_loss[0], 100*d_loss[1], g_loss))

        # If at save interval => save generated image samples
        if epoch % save_interval == 0:
            save_imgs(epoch)
````

**Contexto histórico:**

John Hopfield, físico norteamericano nacido en 1933, dedicó gran parte de su investigación al funcionamiento del cerebro humano, al estudio de sistemas biológicos y al modelado de sistemas neuronales.

Hopfield creía que existia una manera de representar el funcionamiento del cerebro con sistemas digitales. De manera que si se tomaba una tarea biológica elemental y se descomponía en componentes equivalentes (neuronas) y sus conexiones (sinapsis), teniendo en cuenta que el estado del sistema depende del estado actual y de las neuronas vecinas, la tarea se podría adaptar a circuitos integrados usando procesamiento asíncrono paralelo y haciendo uso de la integración de circuitos electrónicos se podian lograr circuitos mucho mas complejos.

El modelo de Hopfield fue formulado a principios de los 80 (1982), como un hito para el renacimiento de las redes neuronales, convirtiendose en un modelo influyente. El primer modelo presentado consistía en un modelo binario cuya funcion de activación era la escalón. Posteriormente diseñó una versión continua con funcion de activación sigmoide.




**¿Qué es la red de Hopfield?**

La red neural de Hopfield es una red recurrente (redes neuronales artificiales donde las conexiones entre nodos forman un gráfico dirigido a lo largo de una secuencia de tiempo) que funciona como un sistema de memoria asociativa que es capaz de almacenar patrones internamente, y reconocerlos incluso a partir de información parcial de los mismos recibida como entrada.

En una red Hopfiel todos los nodos están completamente interconectados, y cada uno de ellos actúa como entrada y como salida, es decir que cada nodo se convierte en una entrada y en una salida para todos los nodos de la red, incluso se puede pensar en que cada nodo es una entrada y salida de sí mismo con un peso de cero. **Funcionamiento Asíncrono o Secuencial**: Se actualiza la salida de una sola neurona en cada iteración, por lo que la salida de la red puede converger a valores diferentes dependiendo del orden en el que se activen las neuronas.

### **2. Aplicaciones de la red**

Las redes monocapa se utilizan típicamente en tareas relacionadas con lo que se conoce como autoasociación; por ejemplo, para regenerar informaciones de entrada que se presentan a la red incompletas o distorsionadas

* **Reconocimiento de imágenes.**
En este tipo de aplicaciones [Koch 86] la Red Hopfield puede reconstruir versiones distorsionadas o parciales de imágenes almacenadas durante la fase de aprendizaje. Para ello sería necesario introducir nuevas reglas de aprendizaje, diferentes de la de Hebb, que tengan implícitas las invarianzas necesarias como para que se pueda llevar a cabo el reconocimiento
deseado.

* **Resolución de problemas de optimización.**
Una de las aplicaciones de las redes neuronales en general y del modelo de Hopfield en particular es la resolución de problemas de optimización complejos, aquellos cuyo tiempo de resolución mediante algoritmos convencionales crece exponencialmente (kN) o factorialmente (N!) con el aumento del tamaño (N) del problema (p. ej. el problema clásico de colocar N reinas en un tablero de ajedrez de NxN cuadros sin que se den jaque, que trataría de encontrar una posición de las piezas de entre las N2!(N2-N)!/N! combinaciones diferentes, que en caso de un tablero real de N=8 serían más de cuatro mil millones de combinacions, para este problema pueden encontrarse algoritmos convencionales con tiempo de resolución proporcionales a N!).

### **3. Arquitectura de la red**

Consiste en una red monocapa compuesta por una cantidad n de neuronas que se encuentran completamente interconectadas en la que cada neurona es realimentada por cada una de las entradas de las otras y la conexión entre neuronas es bidireccional, pero no hay realimentación propia de la neurona (no es autoconcurrente). La salida de la red se obtiene cuando se alcanza un estado estable, es decir el sistema alcaza la convergencia y ya no hay actualizacion de pesos. No obstante, la red de Hopfield no garantiza que el estado estable sea el más cercano al inicial.

La cantidad n de neruonas es igual a la cantidad de elementos del vector que representa el patrón a almacenar.

Los valores de salida de cada neurona son binarios: 0/1 ó -1/+1. En la versión original del modelo ("DH: DISCRETE HOPFIELD") las funciones de activación de las neuronas eran del tipo escalón. Se trataba, por tanto, de una red discreta discreta, con entradas y salidas binarias; sin embargo, posteriormente Hopfield
desarrolló una versión contínua con entradas y salidas analógicas, utilizando neuronas con funciones de activación tipo sigmoidal ("CH: CONTINOUS HOPFIELD").
Cada neurona de la red se encuentra conectada a todas las demás (conexiones laterales), pero no consigo misma (no existen conexiones auto-recurrentes). Además, los pesos asociados a las conexiones entre pares de neuronas son simétricos. Esto significa que el peso de la conexión de una neurona "i" con otra neurona "j" es de igual valor que el de la conexión de la neurona "j" con la "i" (wij = wji).

Estas redes de Hopfield suelen usarse como memoria auto-asociativa: ante la entrada de un patrón x incompleto o perturbado por ruido, la red evoluciona hacia el patrón completo más cercano.

Función para entrenar la red usando la regla de aprendizaje de Hebbian

````python
def train(neu, training_data):
    w = np.zeros([neu, neu])
    for data in training_data:
        w += np.outer(data, data)
    for diag in range(neu):
        w[diag][diag] = 0
    return w
````

````python

def retrieve_pattern(weights, data, steps=10):
    res = np.array(data)

    for _ in range(steps):
        for i in range(len(res)):
            raw_v = np.dot(weights[i], res)
            if raw_v > 0:
                res[i] = 1
            else:
                res[i] = -1
    return res
````

Función para probar la red

````python
def test(weights, testing_data):
    success = 0.0

    output_data = []

    for data in testing_data:
        true_data = data[0]
        noisy_data = data[1]
        predicted_data = retrieve_pattern(weights, noisy_data)
        if np.array_equal(true_data, predicted_data):
            success += 1.0
        output_data.append([true_data, noisy_data, predicted_data])

    return (success / len(testing_data)), output_data
````

En el siguiente ejemplo se declara  una clase (HopfieldNetwork) en la que se definen funciones para entrenar, predecir y calcular los pesos de la red.

````python
# -*- coding: utf-8 -*-
"""
Created on Sun Jul 29 08:40:49 2018
@author: user
"""
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.cm as cm
from tqdm import tqdm

class HopfieldNetwork(object):      
    def train_weights(self, train_data):
        print("Start to train weights...")
        num_data =  len(train_data)
        self.num_neuron = train_data[0].shape[0]        
        # initialize weights
        W = np.zeros((self.num_neuron, self.num_neuron))
        rho = np.sum([np.sum(t) for t in train_data]) / (num_data*self.num_neuron)        
        # Hebb rule
        for i in tqdm(range(num_data)):
            t = train_data[i] - rho
            W += np.outer(t, t)        
        # Make diagonal element of W into 0
        diagW = np.diag(np.diag(W))
        W = W - diagW
        W /= num_data        
        self.W = W 
    
    def predict(self, data, num_iter=20, threshold=0, asyn=False):
        print("Start to predict...")
        self.num_iter = num_iter
        self.threshold = threshold
        self.asyn = asyn        
        # Copy to avoid call by reference 
        copied_data = np.copy(data)        
        # Define predict list
        predicted = []
        for i in tqdm(range(len(data))):
            predicted.append(self._run(copied_data[i]))
        return predicted
    
    def _run(self, init_s):
        if self.asyn==False:
            """
            Synchronous update
            """
            # Compute initial state energy
            s = init_s
            e = self.energy(s)
            
            # Iteration
            for i in range(self.num_iter):
                # Update s
                s = np.sign(self.W @ s - self.threshold)
                # Compute new state energy
                e_new = self.energy(s)                
                # s is converged
                if e == e_new:
                    return s
                # Update energy
                e = e_new
            return s
        else:
            """
            Asynchronous update
            """
            # Compute initial state energy
            s = init_s
            e = self.energy(s)
            
            # Iteration
            for i in range(self.num_iter):
                for j in range(100):
                    # Select random neuron
                    idx = np.random.randint(0, self.num_neuron) 
                    # Update s
                    s[idx] = np.sign(self.W[idx].T @ s - self.threshold)
                
                # Compute new state energy
                e_new = self.energy(s)                
                # s is converged
                if e == e_new:
                    return s
                # Update energy
                e = e_new
            return s
    
    
    def energy(self, s):
        return -0.5 * s @ self.W @ s + np.sum(s * self.threshold)

    def plot_weights(self):
        plt.figure(figsize=(6, 5))
        w_mat = plt.imshow(self.W, cmap=cm.coolwarm)
        plt.colorbar(w_mat)
        plt.title("Network Weights")
        plt.tight_layout()
        plt.savefig("weights.png")
        plt.show()
````

Antes de entrar en detalle de lo que es una red de creencia profunda, primero debemos comprender lo que es una máquina de Boltzmann con restricciones o RBM por su abreviación en el idioma inglés refiriéndose a Restricted Boltzmann Machines.


Una RBM es un tipo de red neuronal estocástica compuesta por una capa de entrada o capa visible, una capa oculta que se encarga de aprender los pesos para lograr el aprendizaje y finalmente, la capa del sesgo que por defecto siempre se encuentra activada. Las RBM son redes neuronales que se generaron a partir de las cadenas de Markov en el campo de la probabilidad y estadística. Un ejemplo de la arquitectura de una RBM es la que se expone en la siguiente imagen:

image.png


Ejemplo de arquitectura de una Restricted Boltzmann Machine [1].


En la arquitectura de una RBM todas las entradas están conectadas a las neuronas que yacen en la capa oculta y la capa del sesgo que se encuentra conectada a la oculta. Para facilitar su comprensión de esta red neuronal, resaltamos el hecho que ni las entradas ni las neuronas en la capa oculta se conectan entre sí, es decir, es una red neuronal que no goza de una cualidad de redundancia que haría más complejo su entrenamiento [1].


Ahora, que hemos explicado de una forma general lo que son las RBM, damos paso a explicar lo que es una red de creencia profunda.


Una red de creencia profunda o Deep Belief Network (DBN) por sus siglas en inglés, no es más que una red conformada por el apilamiento consecutivo de varias RBM con las respectivas características anteriormente expuestas [1].

Es importante destacar que es muy similar a un perceptrón multicapa, pero la diferencia entre ambas redes yace en la manera de entrenarlas. En el caso particular de la DBN, se entrena de forma jerárquica como se haría con las redes convolucionales profundas, permitiéndole aprender en las primeras capas características generales de los datos y en las últimas capas llevar a cabo la integración de esas características aprendidas en las etapas previas [2].


### **1. Razones para utilizar la red**


Algunas de las razones para considerar utilizar DBNs son las siguientes:


* Pueden trabajar con muy pocos datos etiquetados lo que las hace ideales para llevar a cabo tareas de aprendizaje no supervisado.


* Se pueden entrenar en un intervalo de tiempo razonable mediante procesadores gráficos (GPUs).


* Producen resultados sumamente precisos si la comparamos con su versión más superficial que es la RBM.


* Soluciona el problema del desvanecimiento de los gradientes ya que una de las tantas soluciones propuestas para manejar este problema es entrenar de forma jerárquica, es decir, por un nivel a la vez.

### **4. Propiedades e hiperparámetros**

Como se mencionó inicialmente una red de creencia profunda o Deep Belief Network (DBN), se compone de una RBM, por lo cual se listan los principales hiperparámetros de una red RBM: [6]

* **weightInit**: permite ajustar los pesos de los coeficientes con los cuales se entrena la red, unos pesos óptimos generan una clasificación más precisa en el modelo.

* **activationFunction:** se refiere a una de un conjunto de funciones que determinan los umbrales en cada nodo por encima del cual se pasa una señal a través del nodo y por debajo del cual se bloquea. Si un nodo pasa la señal, se "activa".

* **optimizationAlgo:** permite realizar un ajuste de coeficientes con el fin de minimizar el error del modelo.

* **regularization:** permiten implementar regularización de tipo L2, la cual reduce el sobreajuste del modelo, por medio de la penalización en los coeficientes.

* **VisibleUnit / HiddenUnit:** refiere a las capas de una red neuronal. La VisibleUnit, o capa, es la capa de nodos donde entra la entrada, y HiddenUnites la capa donde esas entradas se recombinan en características más complejas. Ambas unidades tienen sus propias llamadas transformaciones, en este caso gaussianas para lo lineal visible y rectificado para lo oculto, que mapean la señal que sale de sus respectivas capas en un nuevo espacio.

* **learningRate/momentum:** ajustan los coeficientes de la red neuronal en cada iteración a medida que corrige el error. Estos dos parámetros ayudan a determinar el tamaño de los pasos que la red baja por el gradiente hacia un óptimo local. Una gran tasa de aprendizaje hará que la red aprenda rápido y quizás sobrepase lo óptimo. Una pequeña tasa de aprendizaje ralentizará el aprendizaje, lo que puede ser ineficiente.

Adicional a los anteriores hiperparametros en una red DBN se encuentra k: 

* **k:** representa el número de veces que ejecuta divergencia contrastante. Cada vez que se ejecuta una divergencia contrastante, es una muestra de la cadena de Markov. Al componer una red de creencias profundas, un valor típico es 1.[6]

En primer lugar debe hacerse la aclaración de que un Autoencoder variacional no es lo mismo que un Autoencoder disperso. Tienen diferentes arquitecturas y diferente funcionamiento a pesar de que su objetivo es el mismo. A continuación se explicara la razón detrás de esta introducción.

Es importante tener en cuenta la estructura de un autoencoder primero. Recordemos que un Autoencoder se compone basicamente de 3 partes: 

<img src="https://d1jnx9ba8s6j9r.cloudfront.net/blog/wp-content/uploads/2018/10/Input-Output-layers-768x736.png" width="400" height="350"><br>
**Imagen 1: ejemplo de arquitectura general de un autoencoder**

Ahora, típicamente el objetivo principal de un Autoencoder es la reducción dimensional de las características principales de un espacio de información (input)... este espacio de información ha resultado en muchos casos ser imágenes o audio decodificado para poder ser procesado por la red neuronal. 

Un Autoencoder tradicional, como el que se ve en la Imágen 1, cumple esta función de reducción dimensional, creando un **<em>espacio de características o espacio latente</em>** en la capa oculta más pequeña, llamada **<em>Bottleneck o cuello de botella</em>**. Este espacio de características esta representado por los pesos del Bottleneck, calculados a partir del entrenamiento de la red neuronal. Finalmente, es importante tener en cuenta que en un Autoencoder tradicional, todas las neuronal de ese espacio de características están activas y por eso es que esta capa de neuronas tiene la menor cantidad de neuronas que el resto de capaz.  

Con relación a un Autoencoder tradicional, tanto el **Sparse autoencoder** como el  **Variational autoencoder** aplican técnicas de regularización que colaboran a una selección más óptima de las características principales del espacio de entrada. Las cuales serán introducidas a continuación: 


### Sparse Autoencoder

La razón por la cual este tipo de Autoencoder puede tener más neuronas en sus capas ocultas que el Autoencoder tradicional es porque la selección de las neuronas del espacio latente se hacen mediante la prodiedad de **sparcity** (dispersión) la cual consiste en una restricción adicional de los pesos en la capa del Bottleneck que hace que solo se activen ciertas neuronas a la hora del entrenamiento de la red neuronal. La intuición basicamente dice que con pesos cercanos a 0, la neurona no se activará, mientras que a medida que el valor del peso se va acercando a 1, la neurona va siendo más activa para la determinación de características en una red neuronal.

Todo esto acerca de los pesos ¿Suena familiar? Por supuesto... es la misma estrategía que usa la regresión lineal multiple (o regresión multinomial) para regularizar los pesos de los terminos del polinomio en el entrenamiento del modelo usando regularización L1 y L2 (Ridge y Lasso). No obstante, la regularización de las funciones L1 y L2 no son las únicas que ayudan a optimizar la tarea de similitud que implica la predicción de valores cercanos a una distribución. 

Puede que lo anterior haya sonado algo confuso, pero recordemos que finalmente lo que quiere la regularización es hacer que una función g(x) de valores predichos, se acerquen a una función f(x) de valores reales. Para esta tarea, nos sirven las funciones de costo que maximizan dicha cercanía, ya sea de una manera estocástica (probabilística) o midiendo la distancia media entre un punto predicho y uno real (como lo hacen Lasso y Ridge).

Es por esta razón que la función de costo usada en el algoritmo de retropropagación para entrenar los Sparse Autoencoders puede ser cualquiera de dichas funciones de regularización.

**<em>En particular, se ha [comprobado](https://www.aclweb.org/anthology/P99-1004.pdf) que una buena elección de la función de penalización de los costos es la divergencia-KL, una explicación detallada de su funcionamiento se encuentra [acá](https://debuggercafe.com/sparse-autoencoders-using-kl-divergence-with-pytorch/).</em>**

A modo de resumen, un Sparse Autoencoder es lo mismo que un autoencoder normal, solo que funciona con una regularización aplicada a su función de costo. Esta regularización puede ser la misma que se ha visto en anteriores modelos, como lo son Ridge y Lasso, no obstante hay más funciones que se ha comprobado que funcionan mejor... como la función de la divergencia KL. 

### Variational Autoencoder

A diferencia de la regularización el el Sparse Autoencoder, la regularización del Variational autoencoder no se encuentra unicamente en su función de costo. Lo interesante de los Variational Autoencoder es que pueden al mismo tiempo usar una función de costo regularizada y traducir las características del espacio latente en una distribución de probabilidad estandarizada. 

Como se mostrará más abajo en la arquitectura, en este tipo de Autoencoder la arquitectura posee en su espacio latente (el que se encuentra en el Bottleneck) no solo un vector de matrices (tensor) con los pesos de cada neurona, sino que traduce esos pesos a otros dos vectores o tensores, uno en representación de la varianza o desviación estandar de la distribución de las carácterísticas, y otro en representación de la media de las características. 

La intuición detrás de este tipo de Autoencoder es que al traducir el espacio latente como una distribución, unicamente se debe tomar una muestra de dicho espacio latente para reconstruir la imagen. Profundizando un poco más acerca del funcionamiento de dicha red se puede dar cuenta que la función de costo de la red neuronal, al aplicar la **divergencia-KL** minimiza la diferencia entre la distribución generada y una distribución Gaussiana o normal estandar (~N(0,1)). 

Para un mejor detalle de cómo funcionan estos Autoencoders se puede ver este video: [Variational Autoencoders](https://www.youtube.com/watch?v=9zKuYvjFFS8)

````python
from tensorflow.keras.layers import Input, Dense, Lambda
from tensorflow.keras.backend import random_normal, exp
from tensorflow.keras.models import Model

# Dimensión datos entradas
input_shape = (28, )

# Tamaño capa oculta intermedia
intermediate_dim = 16

# Batch size
batch_size = 50

# Dimensión de las variables latentes
latent_dim = 2

# Desviación estandard de la función generadora de variables latentes
epsilon_std = 1

# Encoder - Pasa del input a la dimensión latente media y varianza
x = Input(shape=input_shape)
h_encoder = Dense(intermediate_dim, activation='relu')(x)
z_mean = Dense(latent_dim)(x)
z_var = Dense(latent_dim)(x)

# Se generan variables latentes basandose en una distribución gaussiana con media 0 y deviación estandard epsilon_std (hiperparametro)
def sampling(args):
    z_mean, z_var = args
    epsilon = random_normal(shape=(batch_size, latent_dim), mean=0., std=epsilon_std)
    
    return z_mean + exp(z_var) * epsilon
z = Lambda(sampling, output_shape=(latent_dim,))([z_mean, z_var])

# Decoder - Pasa del espacio latente a dimensión de entrada 
h_decoded = Dense(intermediate_dim, activation='relu')(z)
decoder_mean = Dense(input_shape[0], activation='sigmoid')(h_decoded)

# Modelo Autoencoder (end to end)
vae = Model(x, decoder_mean)

# Encoder (Convierte a espacio latente)
encoder = Model(x, z_mean)
````

````python
'''
  Variational Autoencoder (VAE) with the Keras Functional API.
'''

import keras
from keras.layers import Conv2D, Conv2DTranspose, Input, Flatten, Dense, Lambda, Reshape
from keras.layers import BatchNormalization
from keras.models import Model
from keras.datasets import mnist
from keras.losses import binary_crossentropy
from keras import backend as K
import numpy as np
import matplotlib.pyplot as plt

# Load MNIST dataset
(input_train, target_train), (input_test, target_test) = mnist.load_data()

# Data & model configuration
img_width, img_height = input_train.shape[1], input_train.shape[2]
batch_size = 128
no_epochs = 100
validation_split = 0.2
verbosity = 1
latent_dim = 2
num_channels = 1

# Reshape data
input_train = input_train.reshape(input_train.shape[0], img_height, img_width, num_channels)
input_test = input_test.reshape(input_test.shape[0], img_height, img_width, num_channels)
input_shape = (img_height, img_width, num_channels)

# Parse numbers as floats
input_train = input_train.astype('float32')
input_test = input_test.astype('float32')

# Normalize data
input_train = input_train / 255
input_test = input_test / 255

# # =================
# # Encoder
# # =================

# Definition
i       = Input(shape=input_shape, name='encoder_input')
cx      = Conv2D(filters=8, kernel_size=3, strides=2, padding='same', activation='relu')(i)
cx      = BatchNormalization()(cx)
cx      = Conv2D(filters=16, kernel_size=3, strides=2, padding='same', activation='relu')(cx)
cx      = BatchNormalization()(cx)
x       = Flatten()(cx)
x       = Dense(20, activation='relu')(x)
x       = BatchNormalization()(x)
mu      = Dense(latent_dim, name='latent_mu')(x)
sigma   = Dense(latent_dim, name='latent_sigma')(x)

# Get Conv2D shape for Conv2DTranspose operation in decoder
conv_shape = K.int_shape(cx)

# Define sampling with reparameterization trick
def sample_z(args):
  mu, sigma = args
  batch     = K.shape(mu)[0]
  dim       = K.int_shape(mu)[1]
  eps       = K.random_normal(shape=(batch, dim))
  return mu + K.exp(sigma / 2) * eps

# Use reparameterization trick to ....??
z       = Lambda(sample_z, output_shape=(latent_dim, ), name='z')([mu, sigma])

# Instantiate encoder
encoder = Model(i, [mu, sigma, z], name='encoder')
encoder.summary()

# =================
# Decoder
# =================

# Definition
d_i   = Input(shape=(latent_dim, ), name='decoder_input')
x     = Dense(conv_shape[1] * conv_shape[2] * conv_shape[3], activation='relu')(d_i)
x     = BatchNormalization()(x)
x     = Reshape((conv_shape[1], conv_shape[2], conv_shape[3]))(x)
cx    = Conv2DTranspose(filters=16, kernel_size=3, strides=2, padding='same', activation='relu')(x)
cx    = BatchNormalization()(cx)
cx    = Conv2DTranspose(filters=8, kernel_size=3, strides=2, padding='same',  activation='relu')(cx)
cx    = BatchNormalization()(cx)
o     = Conv2DTranspose(filters=num_channels, kernel_size=3, activation='sigmoid', padding='same', name='decoder_output')(cx)

# Instantiate decoder
decoder = Model(d_i, o, name='decoder')
decoder.summary()

# =================
# VAE as a whole
# =================

# Instantiate VAE
vae_outputs = decoder(encoder(i)[2])
vae         = Model(i, vae_outputs, name='vae')
vae.summary()

# Define loss
def kl_reconstruction_loss(true, pred):
  # Reconstruction loss
  reconstruction_loss = binary_crossentropy(K.flatten(true), K.flatten(pred)) * img_width * img_height
  # KL divergence loss
  kl_loss = 1 + sigma - K.square(mu) - K.exp(sigma)
  kl_loss = K.sum(kl_loss, axis=-1)
  kl_loss *= -0.5
  # Total loss = 50% rec + 50% KL divergence loss
  return K.mean(reconstruction_loss + kl_loss)

# Compile VAE
vae.compile(optimizer='adam', loss=kl_reconstruction_loss)

# Train autoencoder
vae.fit(input_train, input_train, epochs = no_epochs, batch_size = batch_size, validation_split = validation_split)

# =================
# Results visualization
# Credits for original visualization code: https://keras.io/examples/variational_autoencoder_deconv/
# (François Chollet).
# Adapted to accomodate this VAE.
# =================
def viz_latent_space(encoder, data):
  input_data, target_data = data
  mu, _, _ = encoder.predict(input_data)
  plt.figure(figsize=(8, 10))
  plt.scatter(mu[:, 0], mu[:, 1], c=target_data)
  plt.xlabel('z - dim 1')
  plt.ylabel('z - dim 2')
  plt.colorbar()
  plt.show()

def viz_decoded(encoder, decoder, data):
  num_samples = 15
  figure = np.zeros((img_width * num_samples, img_height * num_samples, num_channels))
  grid_x = np.linspace(-4, 4, num_samples)
  grid_y = np.linspace(-4, 4, num_samples)[::-1]
  for i, yi in enumerate(grid_y):
      for j, xi in enumerate(grid_x):
          z_sample = np.array([[xi, yi]])
          x_decoded = decoder.predict(z_sample)
          digit = x_decoded[0].reshape(img_width, img_height, num_channels)
          figure[i * img_width: (i + 1) * img_width,
                  j * img_height: (j + 1) * img_height] = digit
  plt.figure(figsize=(10, 10))
  start_range = img_width // 2
  end_range = num_samples * img_width + start_range + 1
  pixel_range = np.arange(start_range, end_range, img_width)
  sample_range_x = np.round(grid_x, 1)
  sample_range_y = np.round(grid_y, 1)
  plt.xticks(pixel_range, sample_range_x)
  plt.yticks(pixel_range, sample_range_y)
  plt.xlabel('z - dim 1')
  plt.ylabel('z - dim 2')
  # matplotlib.pyplot.imshow() needs a 2D array, or a 3D array with the third dimension being of shape 3 or 4!
  # So reshape if necessary
  fig_shape = np.shape(figure)
  if fig_shape[2] == 1:
    figure = figure.reshape((fig_shape[0], fig_shape[1]))
  # Show image
  plt.imshow(figure)
  plt.show()

# Plot results
data = (input_test, target_test)
viz_latent_space(encoder, data)
viz_decoded(encoder, decoder, data)
````

### **3. Arquitectura de la red**

La red consiste únicamente de una capa de entrada (input) y una capa de salida, también denominada <b>capa competitiva</b>. 

* <b>Capa de entrada: </b> contiene tantas neuronas como variables predictoras haya. 
* <b>Capa competitiva: </b> es una cuadrícula topológica de neuronas, de baja dimensión (por lo general tiene 2 dimensiones). El número de neuronas en esta capa depende del problema, pudiendo variar de unas pocas a varios miles de neuronas. 

A continuación, presentamos un ejemplo de la arquitectura de la red:

![texto alternativo](https://miro.medium.com/max/420/0*_NjtJEOILJokl29R.png)

En cuanto a la conectividad entre las dos capas, cada neurona de la capa de entrada está conectada a todas las neuronas de la capa competitiva. Así pues, un vector de N (# de variables) dimensiones se le asigna a cada neurona de la capa competitiva.

El objetivo del algoritmo es calcular el modelo que mejor describe las observaciones. Estos modelos se organizan en la cuadrícula de la capa de salida, de modo que modelos similares entre sí son más cercanos. Este mapeo auto-organizativo mantiene una relación de vecindad entre la distribución original de los datos N-dimensionales y la cuadrícula topológica de baja dimensión. 

#### ¿Por qué capa *competitiva*?

El nombre de esta capa proviene del algoritmo que aplica la red, el cual tiene el objetivo de calcular el vector de pesos para cada neurona de salida. A grandes rasgos, el algoritmo opera de la siguiente manera:

1. Inicialización (aleatoria) de los pesos.

2. Proceso iterativo (optimización de pesos):
    1. Se escoge al azar una observación X
    2. Se calcula la distancia (por lo general Euclidiana) entre la posición N-dimensional de la observación X y el vector de pesos asociado a cada neurona de la capa de salida.
    3. La neurona cuyo vector de pesos se encuentre más cercano a la observación, se le llama neurona <b> ganadora </b>.    
    4. Los vecinos topológicos de la neurona ganadora se acercan a la posición de la observación.    
    5. Se vuelve al paso A escogiendo otra observación.
    
Por tanto, las neuronas "compiten" por representar observaciones.

Sin embargo, esta es la implementación de la DESOM (Deep Embedded SOM). No obstante, ese repositorio de Github que contiene el código oficial de la implementación de Keras de esta variación contiene el script llamado *kerasom.py* que contiene la forma estándar de una red SOM usando Keras.

El código contiene código para visualizaciones de los resultados, funciones de pérdida, y más funciones. El que concierne a la implementación de la red es: 

````python
class Kerasom:
    """
    Kerasom model (standard SOM in Keras)
    # Example
        ```
        kerasom = Kerasom(input_dim=784, map_size=(10,10))
        ```
    # Arguments
        input_dim: input vector dimension
        map_size: tuple representing the size of the rectangular map. Number of prototypes is map_size[0]*map_size[1]
    """

    def __init__(self, input_dim, map_size):
        self.input_dim = input_dim
        self.map_size = map_size
        self.n_prototypes = map_size[0]*map_size[1]
        self.input = None
        self.model = None
    
    def initialize(self):
        """
        Create and compile SOM model
        """
        self.input = Input(shape=(self.input_dim,), name='input')
        som_layer = SOMLayer(self.map_size, name='SOM')(self.input)
        self.model = Model(inputs=self.input, outputs=som_layer)
    
    @property
    def prototypes(self):
        """
        Returns SOM code vectors
        """
        return self.model.get_layer(name='SOM').get_weights()[0]

    def compile(self, optimizer):
        """
        Compile Kerasom model
        # Arguments
            optimizer: optimization algorithm
        """
        self.model.compile(loss=som_loss, optimizer=optimizer)
    
    def load_weights(self, weights_path):
        """
        Load pre-trained weights of Kerasom model
        """
        self.model.load_weights(weights_path)

    def init_som_weights(self, X):
        """
        Initialize with a sample w/o remplacement of encoded data points.
        """
        sample = X[np.random.choice(X.shape[0], size=self.n_prototypes, replace=False)]
        self.model.get_layer(name='SOM').set_weights([sample])

    def predict(self, x):
        """
        Predict best-matching unit using the output of SOM layer
        # Arguments
            x: data point
        # Return
            index of the best-matching unit
        """
        d = self.model.predict(x, verbose=0)
        return d.argmin(axis=1)

    def map_dist(self, y_pred):
        """
        Calculate pairwise Manhattan distances between cluster assignments and map prototypes (rectangular grid topology)
        
        # Arguments
            y_pred: cluster assignments, numpy.array with shape `(n_samples,)`
        # Return
            pairwise distance matrix (map_dist[i,k] is the distance on the map between assigned cell of data point i and cell k)
        """
        labels = np.arange(self.n_prototypes)
        tmp = np.expand_dims(y_pred, axis=1)
        d_row = np.abs(tmp-labels)//self.map_size[1]
        d_col = np.abs(tmp%self.map_size[1]-labels%self.map_size[1])
        return d_row + d_col

    @staticmethod
    def neighborhood_function(d, T, neighborhood='gaussian'):
        """
        SOM neighborhood function (gaussian neighborhood)
        # Arguments
            d: distance on the map
            T: temperature parameter
        """
        if neighborhood == 'gaussian':
            return np.exp(-(d**2)/(T**2))
        elif neighborhood == 'window':
            return (d <= T).astype(np.float32)
    
    def fit(self, X_train, y_train=None,
            X_val=None, y_val=None,
            iterations=10000,
            som_iterations=10000,
            eval_interval=10,
            save_epochs=5,
            batch_size=256,
            Tmax=10,
            Tmin=0.1,
            decay='exponential',
            save_dir='results/tmp'):
        """
        Training procedure
        # Arguments
           X_train: training set
           y_train: (optional) training labels
           X_val: (optional) validation set
           y_val: (optional) validation labels
           iterations: number of training iterations
           som_iterations: number of iterations where SOM neighborhood is decreased
           eval_interval: evaluate metrics on training/validation batch every eval_interval iterations
           save_epochs: save model weights every save_epochs epochs
           batch_size: training batch size
           Tmax: initial temperature parameter
           Tmin: final temperature parameter
           decay: type of temperature decay ('exponential' or 'linear')
           save_dir: path to existing directory where weights and logs are saved
        """
        save_interval = X_train.shape[0] // batch_size * save_epochs # save every save_epochs epochs
        print('Save interval:', save_interval)

        # Logging file
        logfile = open(save_dir + '/kerasom_log.csv', 'w')
        fieldnames = ['iter', 'T', 'Lsom', 'Lkm', 'Ltop', 'quantization_err', 'topographic_err']
        if X_val is not None:
            fieldnames += ['Lsom_val', 'Lkm_val', 'Ltop_val', 'quantization_err_val', 'topographic_err_val']
        if y_train is not None:
            fieldnames += ['acc', 'pur', 'nmi', 'ari']
        if y_val is not None:
            fieldnames += ['acc_val', 'pur_val', 'nmi_val', 'ari_val']
        logwriter = csv.DictWriter(logfile, fieldnames)
        logwriter.writeheader()

        # Set and compute some initial values
        index = 0
        if X_val is not None:
            index_val = 0
        T = Tmax

        for ite in range(iterations):
            # Get training and validation batches
            if (index + 1) * batch_size > X_train.shape[0]:
                X_batch = X_train[index * batch_size::]
                if y_train is not None:
                    y_batch = y_train[index * batch_size::]
                index = 0
            else:
                X_batch = X_train[index * batch_size:(index + 1) * batch_size]
                if y_train is not None:
                    y_batch = y_train[index * batch_size:(index + 1) * batch_size]
                index += 1
            if X_val is not None:
                if (index_val + 1) * batch_size > X_val.shape[0]:
                    X_val_batch = X_val[index_val * batch_size::]
                    if y_val is not None:
                        y_val_batch = y_val[index_val * batch_size::]
                    index_val = 0
                else:
                    X_val_batch = X_val[index_val * batch_size:(index_val + 1) * batch_size]
                    if y_val is not None:
                        y_val_batch = y_val[index_val * batch_size:(index_val + 1) * batch_size]
                    index_val += 1

            # Compute cluster assignments for batches
            d = self.model.predict(X_batch)
            y_pred = d.argmin(axis=1)
            if X_val is not None:
                d_val = self.model.predict(X_val_batch)
                y_val_pred = d_val.argmin(axis=1)

            # Update temperature parameter
            if ite < som_iterations:
                if decay == 'exponential':
                    T = Tmax*(Tmin/Tmax)**(ite/(som_iterations-1))
                elif decay == 'linear':
                    T = Tmax - (Tmax-Tmin)*(ite/(som_iterations-1))

            # Compute topographic weights batches
            w_batch = self.neighborhood_function(self.map_dist(y_pred), T, neighborhood='gaussian')
            if X_val is not None:
                w_val_batch = self.neighborhood_function(self.map_dist(y_val_pred), T, neighborhood='gaussian')

            # Train on batch
            loss = self.model.train_on_batch(X_batch, w_batch)

            if ite % eval_interval == 0:
                # Initialize log dictionary
                logdict = dict(iter=ite, T=T)

                # Evaluate losses and metrics
                print('iteration {} - T={}'.format(ite, T))
                logdict['Lsom'] = loss
                logdict['Lkm'] = kmeans_loss(y_pred, d)
                logdict['Ltop'] = loss - logdict['Lkm']
                logdict['quantization_err'] = quantization_error(d)
                logdict['topographic_err'] = topographic_error(d, self.map_size)
                print('[Train] - Lsom={:f} (Lkm={:f}/Ltop={:f})'.format(logdict['Lsom'], logdict['Lkm'], logdict['Ltop']))
                print('[Train] - Quantization err={:f} / Topographic err={:f}'.format(logdict['quantization_err'], logdict['topographic_err']))
                if X_val is not None:
                    val_loss = self.model.test_on_batch(X_val_batch, [X_val_batch, w_val_batch])
                    logdict['Lsom_val'] = val_loss
                    logdict['Lkm_val'] = kmeans_loss(y_val_pred, d_val)
                    logdict['Ltop_val'] = val_loss - logdict['Lkm_val']
                    logdict['quantization_err_val'] = quantization_error(d_val)
                    logdict['topographic_err_val'] = topographic_error(d_val, self.map_size) 
                    print('[Val] - Lsom={:f} (Lkm={:f}/Ltop={:f})'.format(logdict['Lsom_val'], logdict['Lkm_val'], logdict['Ltop_val']))
                    print('[Val] - Quantization err={:f} / Topographic err={:f}'.format(logdict['quantization_err_val'], logdict['topographic_err_val']))

                # Evaluate the clustering performance using labels
                if y_train is not None:
                    logdict['acc'] = cluster_acc(y_batch, y_pred)
                    logdict['pur'] = cluster_purity(y_batch, y_pred)
                    logdict['nmi'] = metrics.normalized_mutual_info_score(y_batch, y_pred)
                    logdict['ari'] = metrics.adjusted_rand_score(y_batch, y_pred)
                    print('[Train] - Acc={:f}, Pur={:f}, NMI={:f}, ARI={:f}'.format(logdict['acc'], logdict['pur'], logdict['nmi'], logdict['ari']))
                if y_val is not None:
                    logdict['acc_val'] = cluster_acc(y_val_batch, y_val_pred)
                    logdict['pur_val'] = cluster_purity(y_val_batch, y_val_pred)
                    logdict['nmi_val'] = metrics.normalized_mutual_info_score(y_val_batch, y_val_pred)
                    logdict['ari_val'] = metrics.adjusted_rand_score(y_val_batch, y_val_pred)
                    print('[Val] - Acc={:f}, Pur={:f}, NMI={:f}, ARI={:f}'.format(logdict['acc_val'], logdict['pur_val'], logdict['nmi_val'], logdict['ari_val']))
                    
                logwriter.writerow(logdict)

            # Save intermediate model
            if ite % save_interval == 0:
                self.model.save_weights(save_dir + '/kerasom_model_' + str(ite) + '.h5')
                print('Saved model to:', save_dir + '/kerasom_model_' + str(ite) + '.h5')

        # Save the final model
        logfile.close()
        print('saving model to:', save_dir + '/kerasom_model_final.h5')
        self.model.save_weights(save_dir + '/kerasom_model_final.h5')

````

### **5. Propiedades e hiperparametros**

En esta sección se describirán las propiedades del proceso clave de autoorganización, que es el seguido por esta metodología el cual está conformado por cuatro fases. Por otro lado al describir estas fases, se realizará énfasis en los hiperparámetros.

#### *1. Inicialización:*

Se asignan pesos de manera aleatoria.

#### *2. Competencia:*

Este es el paso mediante el cual se garantiza que la entrada, la cual corresponde a un espacio continuo, sea transformado a un espacio discreto conformado por neuronas.

![texto alternativo](https://media-exp1.licdn.com/dms/image/C4D12AQHnL49dLOUGCQ/article-inline_image-shrink_1000_1488/0?e=1602115200&v=beta&t=SUzYvSvU8qF9L0j5tXnClTVo00KFPT3OmCXtYIcDQPQ)

#### *3. Cooperación:*

La activación de una neurona genera una excitación mayor en las que se encuentran más cercanas a esta, lo cual se permite medir la distancia y determinar aquellas que se encuentran en el vecindario, al ser un vecindario topológico presenta caída ante el aumento de la distancia.

La función que define el vecindario topológico se caracteriza por:

1. Presenta su máximo valor en la neurona ganadora.
2. Es una función simétrica, esta simetría se presenta alrededor de la neurona.
3. El límite cuando la distancia tiende a infinito es 0.
4. Es invariante en la translación (independiente de la posición de la nuerona ganadora)
5. La función depende de la variable $\sigma$, cuyo valor delimita el tamaño del vecindario.

**Hiperparámetros**

El tamaño del vecindario $\sigma$, depende de dos de estos:
1. $\tau_0$ asociado al tiempo.
2. $\sigma_0$ asociado al radio inicial de la vecindad.


### *4. Adaptación:*

El vecindario topológico logra mediante las iteraciones adaptar los pesos y por ende los vecinos de las neuronas.


![texto alternativo](https://media-exp1.licdn.com/dms/image/C4D12AQFXOr5bC3w1Ag/article-inline_image-shrink_1000_1488/0?e=1602115200&v=beta&t=4iCDi5Ihi-aCXxRM24aGmDu1n62IG5joVQGM2owUqKk)

2. La neurona se desplaza hacía el punto blaco, y la siguen las neuronas conectadas. 
3. Luego de muchas iteraciones la red se adapta a la topología de los datos originales.

#### **6.3. Modelamiento**

El SOM reajusta los pesos de la red de acuerdo con la cantidad de iteraciones que se realicen. A continuación, se define una lista con diferentes valores de iteraciones con el fin de evidenciar los resultados del SOM en cada iteración:

##### **6.2.1. TensorFlow**

Para realizar la implementación del SOM empleando TensorFlow, se define la siguiente clase encargada de ejecutar las operaciones internas de la red para obtener el comportamiento del SOM:

````python
class SOM(object):

    # To check if the SOM has been trained
    trained = False

    def __init__(self, m, n, dim, n_iterations=100, alpha=None, sigma=None):

        # Assign required variables first
        self.m = m; self.n = n
        if alpha is None:
            alpha = 0.2
        else:
            alpha = float(alpha)
        if sigma is None:
            sigma = max(m, n) / 2.0
        else:
            sigma = float(sigma)
        self.n_iterations = abs(int(n_iterations))

        self.graph = tf.Graph()

        with self.graph.as_default():

            # To save data, create weight vectors and their location vectors

            self.weightage_vects = tf.Variable(tf.random_normal( [m * n, dim]) )

            self.location_vects = tf.constant(np.array(list(self.neuron_locations(m, n))))

            # Training inputs

            # The training vector
            self.vect_input = tf.placeholder("float", [dim])
            # Iteration number
            self.iter_input = tf.placeholder("float")

            # Training Operation  # tf.pack result will be [ (m*n),  dim ]

            bmu_index = tf.argmin(tf.sqrt(tf.reduce_sum(
                tf.pow(tf.subtract(self.weightage_vects, tf.stack(
                    [self.vect_input for _ in range(m * n)])), 2), 1)), 0) 
                    

            slice_input = tf.pad(tf.reshape(bmu_index, [1]), np.array([[0, 1]]))
            bmu_loc = tf.reshape(tf.slice(self.location_vects, 
                                    slice_input, tf.constant(np.array([1, 2]))), [2])

            # To compute the alpha and sigma values based on iteration number
            learning_rate_op = tf.subtract(1.0, tf.div(self.iter_input, self.n_iterations))
            alpha_op = tf.multiply(alpha, learning_rate_op)
            sigma_op = tf.multiply(sigma, learning_rate_op)

            # learning rates for all neurons, based on iteration number and location w.r.t. BMU.
            bmu_distance_squares = tf.reduce_sum(tf.pow(tf.subtract(
                self.location_vects, tf.stack( [bmu_loc for _ in range(m * n)] ) ) , 2 ), 1)

            neighbourhood_func = tf.exp(tf.negative(tf.div(tf.cast(
                bmu_distance_squares, "float32"), tf.pow(sigma_op, 2))))
            learning_rate_op = tf.multiply(alpha_op, neighbourhood_func)

            # Finally, the op that will use learning_rate_op to update the weightage vectors of all neurons
            learning_rate_multiplier = tf.stack([tf.tile(tf.slice(
                learning_rate_op, np.array([i]), np.array([1])), [dim]) for i in range(m * n)] )

            ### Strucutre of updating weight ###
            ### W(t+1) = W(t) + W_delta ###
            ### wherer, W_delta = L(t) * ( V(t)-W(t) ) ###

            # W_delta = L(t) * ( V(t)-W(t) )
            weightage_delta = tf.multiply(
                learning_rate_multiplier,
                tf.subtract(tf.stack([self.vect_input for _ in range(m * n)]), self.weightage_vects))

            # W(t+1) = W(t) + W_delta
            new_weightages_op = tf.add(self.weightage_vects, weightage_delta)

            # Update weightge_vects by assigning new_weightages_op to it.
            self.training_op = tf.assign(self.weightage_vects, new_weightages_op)

            self.sess = tf.Session()
            init_op = tf.global_variables_initializer()
            self.sess.run(init_op)

    def neuron_locations(self, m, n):

        for i in range(m):
            for j in range(n):
                yield np.array([i, j])

    def train(self, input_vects):

        # Training iterations
        for iter_no in range(self.n_iterations):
            # Train with each vector one by one
            for input_vect in input_vects:
                self.sess.run(self.training_op, 
                        feed_dict={self.vect_input: input_vect, self.iter_input: iter_no})

        # Store a centroid grid for easy retrieval later on
        centroid_grid = [[] for i in range(self.m)]
        self.weightages = list(self.sess.run(self.weightage_vects))
        self.locations = list(self.sess.run(self.location_vects))
        for i, loc in enumerate(self.locations):
            centroid_grid[loc[0]].append(self.weightages[i])

        self.centroid_grid = centroid_grid

        self.trained = True

    def get_centroids(self):

        if not self.trained:
            raise ValueError("SOM not trained yet")
        return self.centroid_grid

    def map_vects(self, input_vects):

        if not self.trained:
            raise ValueError("SOM not trained yet")

        to_return = []
        for vect in input_vects:
            min_index = min( [i for i in range(len(self.weightages))], 
                            key=lambda x: np.linalg.norm(vect - self.weightages[x]) )
            to_return.append(self.locations[min_index])

        return to_return
````

**Una Gated Recurent Unit** (GRU) es una versión mejorada de una red neuronal recurrente. Debido a que está soluciona el problema de **vanishing gradient problem**, presente en las redes recurrentes tradicionales, haciendo uso de la técnica **update gate and reset gate**. La cual consiste en tener dos vectores los cuales deciden  que se pasa como ouput.  Una característica especial de ests mismos es que pueden retener información del pasado. 

### **1. Razones para utilizar la red**
Este tipo de redes neuronales son usadas cuando los datos de entrada son de tipo secuenciales. Un dato secuencial se caracteriza por presentar conexiones importantes identificables, o evidentes, sólo cuando el dato a sido recorrido completamente. Para ilustrar mejor la idea, consideremos el siguiente ejemplo, Supongamos que deseamos entrenar un clasificador que identifique el estilo gramatical de Yoda (un ejemplo simple y poco creativo), ej:

Vemos que las frases son idénticas, tanto en las palabras que utiliza como en el significado, por tanto usar algo basado en CountVectorizer o TFID seguramente no arrojará diferencias. Sin embargo, lo importante está en el órden de las palabras. La secuencia de estos datos es importante. Las GRU se enfocan en ellos y funcionan utilizando dos compuertas, una de actualización y otra de reseteo. La última se encarga de definir cuando un dato tiene información importante y debe mantenerse en memoria y cuando debe descartarse.

### **3. Arquitectura de la red**

![Arquitectura de la red](https://miro.medium.com/max/700/1*7oE-4Wg6bZ7u8yDf5cjJPA.png)



> Imagen de:  https://towardsdatascience.com/understanding-gru-networks-2ef37df6c9be

La arquitectura de la red es similar a la arquitectura de un RNN, con la diferencia que implementa una puerta de update y una puerta de reset.

#### Update gate

El input en el tiempo $t$ de la red se multiplica por el vector de pesos, al igual que se hace con el vector $h_{t-1}$. Posteriormente se le aplica la función sigmoide al resultado. El update gate nos permite conocer que tanto del pasado se pone en el futuro.

#### Reset gate

La operación matemática es la misma que la de la update gate, con la diferencia de que el resultado de la función de sigmoide va a otra parte de la red.

Ahora, el resultado de la operación de la reset gate se le aplica el producto Hadamard con el vector $h_{t-1}$, luego le sumamos el vector de la entrada ajustado a los pesos y finalmente aplicamos tangente hiperbólica. La función completa de la reset gate se ve así: $h_t' = tanh(Wx_i + r_i \circ Uh_{t-1})$

Finalmente, el vector $h_{i}$ se calcula usando la siguiente fórmula.

$h_{t} = z_{t} \circ h_{t-1} + (1 - z_{t}) \circ h_t'$. El $z_t$ es el resultado de la update gate.

### **4. La red en Keras**

````python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras import Model
````

````python
#Tomado de https://towardsdatascience.com/implementation-of-rnn-lstm-and-gru-a4250bf6c090

''' Crea GRU.
      rnn_nb: Número de nodos RNN. El tamaño del arreglo es el número de capas RNN.
      fc_nb: Número de nodos FC. El tamaño del arreglo es el número de capas FC.
      Tienen que definir dropout_rate antes de invocar el método. 
'''
def get_model(_rnn_nb, _fc_nb):
  spec_start = layers.Input((256, 256))
  spec_x = spec_start
  for _r in _rnn_nb:
    spec_x = layers.GRU(_r, activation='tanh', dropout=dropout_rate, recurrent_dropout=dropout_rate, return_sequences=True)(spec_x)

  for _f in _fc_nb:
    spec_x = layers.TimeDistributed(layers.Dense(_f))(spec_x)
    spec_x = layers.Dropout(dropout_rate)(spec_x)

  spec_x = layers.TimeDistributed(layers.Dense(10))(spec_x)
  out = layers.Activation('sigmoid', name='strong_out')(spec_x)

  _model = Model(inputs=spec_start, outputs=out)
  _model.compile(optimizer='Adam', loss='binary_crossentropy',metrics = ['accuracy'])
  _model.summary()
  return _model

rnn_nb = [32, 32]
fc_nb = [32]
dropout_rate = 0.5

get_model(rnn_nb, fc_nb)
````
